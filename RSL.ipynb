```python
# Importando as bibliotecas pandas, matplolib e numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Iportando o Arquivo CSV para a análise dos dados, 
#Abaixo será gerado um DataFrame para a manipulação usando o Pandas

df = pd.read_csv('RSL_BDTD.csv')

df
```

```python
# Suponha que você já tenha um DataFrame chamado 'df' com as colunas 'types' e 'publication Dates'

# Converta a coluna 'publication Dates' para o formato de data
df['publicationDates'] = pd.to_datetime(df['publicationDates'])

# Crie um DataFrame separado para cada tipo ('masterThesis' e 'doctoralThesis')
master_thesis_df = df[df['types'] == 'masterThesis']
doctoral_thesis_df = df[df['types'] == 'doctoralThesis']

# Agrupe os dados por data e conte a ocorrência de cada tipo
master_thesis_count = master_thesis_df.groupby('publicationDates').size()
doctoral_thesis_count = doctoral_thesis_df.groupby('publicationDates').size()

# Plote o gráfico de linha
plt.figure(figsize=(12, 6))
plt.plot(master_thesis_count.index, master_thesis_count.values, label='masterThesis', marker='o')
plt.plot(doctoral_thesis_count.index, doctoral_thesis_count.values, label='doctoralThesis', marker='o')

# Defina os rótulos dos eixos e o título do gráfico
plt.xlabel('Data de Publicação')
plt.ylabel('Contagem')
plt.title('Evolução da Contagem de masterThesis e doctoralThesis por Data de Publicação')

# Adicione uma legenda
plt.legend()

# Mostre o gráfico
plt.grid()
plt.show()
```

```python
from wordcloud import WordCloud

# Suponhamos que você tenha um DataFrame chamado 'df_ref' com um campo 'Abstract'.
# Vamos converter todos os valores da coluna 'Abstract' em strings.

df_ref['Title'] = df_ref['Title'].astype(str)

# Combine todos os textos em um único string
text = ' '.join(df_ref['Title'])

# Crie a nuvem de palavras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Exiba a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
